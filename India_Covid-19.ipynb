```python
Indian_Cord=pd.read_excel(r"C:\Users\harsh\OneDrive\Desktop\Projects\Covid 19\Indian Coordinates.xlsx")
```

```python
Indian_Cord
```

```python
# Now we merge the lat and long with the df(the original data of corona cases) 
df_full=pd.merge(Indian_Cord,df,on='Name of State / UT')     #on is similar to where condition in join
df_full
```

```python
#Use folium library to make the maps
map=folium.Map(location=[20,78],zoom_start=4.5,tiles='stamenterrain')     #location consists of lat and long of India on map, zoom_start is like a bit zoom version
                                    #that concentartes on India more tiles is the way we want the map to be can find it online on folium map tiles
    
for lat,long,value,name in zip(df_full['Latitude'],df_full['Longitude'],df_full['Total Cases'],df_full['Name of State / UT']):
    folium.CircleMarker([lat,long], radius=0.8*value, popup=('<strong>State</strong>: '+str(name).capitalize()+'<br>'
                                                             '<strong>Total Cases</strong>: '+str(value)+'<br>'),color='red',fill_color='red',fill_opacity=0.3).add_to(map)
    
    
    #0.8 in radius is to make the size of circle bigger to view and the 0.3 in opacity is 30% filling
```
